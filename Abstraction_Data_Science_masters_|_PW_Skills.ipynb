```python
import abc
class pwskills(abc.ABC):

    @abc.abstractmethod
    def student_detials(self):
        pass

    @abc.abstractmethod
    def student_assignemnt(self):
       pass

    @abc.abstractmethod
    def student_mark(self):
        pass
```

```python
class detils(pwskills):

    def student_detials(self):
        return  "student are very good parson"

    def student_assignemnt(self):
        return " your assignemnt are you creat code in python"
#isme abstractmethod ko nahi follow ka rwha hai isle error bata rahai
```

```python
sajan =  detils()
sajan.student_detials()
```
