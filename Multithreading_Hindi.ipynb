```python
import threading
import urllib.request

def file_downloder(url,filename):
    urllib.request.urlretrieve(url,filename)
```

```python
import time
```

```python


def test2(id):
    for i in range(10):
        print("it's id %d and parrting %d %s" %(id,i,time.ctime()),flush=True)
        time.sleep(1)
```

```python
test2(3)
```

```python
thread1 = [threading.Thread( target = test2 , args = (i,)) for i in range(3)]
```

```python
for i in thread1:
    i.start()
```

```python
shered_ver = 0
lock_ver = threading.Lock()

def test3(id):
    global shered_ver
    with lock_ver:
        shered_ver = shered_ver + 1
        print("test3 is id %d inceased with the shered_ver by %s" %(id,shered_ver))
```

```python
thread3 = [threading.Thread( target = test3 , args = (i,)) for i in range(3)]
```

```python
for r in thread3:
    r.start()
```
