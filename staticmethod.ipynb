```python
class pwskills1:


    def student_details(self,name,emil_id,mobile_number):
        print(name,emil_id,mobile_number)


    @staticmethod
    def mentor_class(list_mentor):
        print(list_mentor)

    def mentor(self,list_mentor):
        print(list_mentor)
```

```python
pwskills1.mentor_class(["sajan",'kuma'])
```

```python
class pwskills2:


    def student_details(self,name,emil_id,mobile_number):
        print(name,emil_id,mobile_number)


    @staticmethod
    def mentor_class(list_mentor):
        print(list_mentor)


    @classmethod
    def class_name(cls):
        cls.mentor_class(["sajan","kumar"])

    def mentor(self,list_mentor):
        print(list_mentor)
```

```python
pwskills2.class_name())
```

```python
class pwskills3:


    def student_details(self,name,emil_id,mobile_number):
        print(name,emil_id,mobile_number)


    @staticmethod
    def mentor_gmail_id(gmail_id_mentor):
        print(gmail_id_mentor)

    @staticmethod
    def mentor_class(list_mentor):

        pwskills3.mentor_gmail_id("sajan$123")
        print(list_mentor)




    @classmethod
    def class_name(cls):
        cls.mentor_class(["sajan","kumar"])

    def mentor(self,list_mentor):
        print(list_mentor)
        self.mentor_class(["banty","kumar"])
```
