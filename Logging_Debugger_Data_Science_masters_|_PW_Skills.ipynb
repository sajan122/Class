```python
l = [1,2,3,34,4,[2,3,4],"sajan","kumar"]
```

```python
l1_int = []
l2_str = []
logging.info("start the code")

for i in l:
    logging.info("the list element {}".format(i)) # 1 method you can use format
                        #ya par ap str() f" ka use kar sakete ho
                        #jo thok lege you can use
    if type(i) == list:
        logging.info(" i found list" + str(i)) #2 method

        for j in i:

                logging.info(f" i am  check list inside list element{j}") # 3 method you can use anyone
                if type(j) == int:
                    logging.info(f"i found int is {j}")
                    l1_int.append(j)
                    logging.info(f" than append {l1_int}")  #"than append {}".format(l1_int)

                else:
                    if type(j) == str:
                        logging.info(f"i found str is {j}")
                        l2_str.append(j)
                        logging.info(f" than append {l2_str}")

    if type(i) == int:
        logging.info(f"i found int is {i}")

        l1_int.append(i)
        logging.info(f" than append {l1_int}")
    else:
        if type(i) == str:
            logging.info(f"i found str is {i}")
            l2_str.append(i)
            logging.info(f" than append {l2_str}")

logging.info(f"my final result {l1_int} and {l2_str}")
```
