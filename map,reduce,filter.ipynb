```python
from functools import reduce
```

```python
l = [2,3,4,5,6]
```

```python
reduce(lambda x,y : x+y ,l)
```

```python
reduce(lambda x,y,z : x+y+z ,l)
```

```python
reduce(lambda x,y : x+y ,[])
```

```python
reduce(lambda x,y : x+y ,[1])
```

```python
reduce(lambda x,y : x*y ,l)
```

```python
l  = [1,2,7,6,8]
reduce(lambda x ,y : x if x>y else y,l)
```
