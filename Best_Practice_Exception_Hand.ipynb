```python
#don't use be genral excption

try:
    10/0
except Exception as e:
    print(e)
```

```python
#use always specific exception
try:
    10/0
except ZeroDivisionError as e:
    print(e) #it's will be right
```

```python
# always use logging(log)
#don't be use print
import logging
logging.basicConfig(filename = "txt.log", level = logging.ERROR)
try:
    10/0
except ZeroDivisionError as e:
    logging.error(f"never will be excute any number {e}")
```

```python
#always avoid a write multiple exception(except)
try:
    10/0
except TypeError as e:
    print(e)

except IndexError as e:
    print(e)

except KeyError as e:
    print(e)
#ek hi lekho avoid mulitiple except
except ZeroDivisionError as e:
    print("never will be excute any number",e)
```
