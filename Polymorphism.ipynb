```python
class data_science:#note data science is eis taharaha ni likh sate hai isme koi gep nahi ho sata hai
    def syllabus(self):
        print("this is syllabus for data science")
```

```python
class web_dev:
    def syllabus(self):
        print("this is syllabilus for web dev")
```

```python
def class_parcer(class_obj):
    for i in class_obj:
        i.syllabus()
```

```python
data_science = data_science()
```

```python
web_dev = web_dev()
```

```python
 class_obj = [data_science ,web_dev]
```

```python
class_parcer(class_obj)
```
