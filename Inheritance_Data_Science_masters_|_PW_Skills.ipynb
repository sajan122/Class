```python
class class1:

    def test_meth1(self):
        return "this is a meth class1"
```

```python
class class2(class1):

    def test_meth2(self):
        return "this is a meth class 2"
```

```python
class class3(class2):
    pass
```

```python
sajan = class3()
```

```python
sajan.test_meth1()
```

```python
sajan.test_meth2()
```
